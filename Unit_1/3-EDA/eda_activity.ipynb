### Activity 3

Refer to the file files_for_activities/merged_clean_ver2.csv for this exercise.

- Load data from the merged_clean_ver2.csv file.
- Plot Pearson Correlation Heatmap.
- Plot Spearman Correlation Heatmap.
- Plot Kendall Correlation Heatmap.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('merged_clean_ver2.csv')
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(data.corr(method='pearson'), annot=True)
plt.show()
```

```python
sns.heatmap(data.corr('pearson'), annot=True)
plt.show()
```

```python
sns.heatmap(data.corr(method='spearman'), annot=True)
plt.show()
```

```python
sns.heatmap(data.corr(method='kendall'), annot=True)
plt.show()
```
